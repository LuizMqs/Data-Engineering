# Total population of macro regions


## 2010

```python
total_population_of_macro_regions_2010 = df_year_2010.groupby("regiao")["pop_ibge"].sum().reset_index()
total_population_of_macro_regions_2010 = total_population_of_macro_regions_2010.rename(columns={'regiao':"macro-regiao","pop_ibge":'populacao-total-2010'})
total_population_of_macro_regions_2010['macro-regiao'] = total_population_of_macro_regions_2010['macro-regiao'].replace('Centro Oeste', 'Centro-Oeste')
total_population_of_macro_regions_2010_asc = total_population_of_macro_regions_2010.sort_values(by='populacao-total-2010', ascending=True)

total_population_of_macro_regions_2010_asc
```

## 2022

```python
total_population_of_macro_regions_2022 = df_year_2022.sort_values(by='populacao-total', ascending=True)
total_population_of_macro_regions_2022 = total_population_of_macro_regions_2022.rename(columns={"populacao-total":"populacao-total-2022"})
total_population_of_macro_regions_2022
```



## Total population merge from 2010 to 2022

```python
total_population_of_macro_regions_2010_and_2022 = pd.merge(total_population_of_macro_regions_2010, total_population_of_macro_regions_2022, on="macro-regiao")
total_population_of_macro_regions_2010_and_2022
```

## Growth rate between 2010 and 2022


```python
total_population_of_macro_regions_2010_and_2022['taxa-variacao-ano'] = ((total_population_of_macro_regions_2010_and_2022['populacao-total-2022'] / total_population_of_macro_regions_2010_and_2022['populacao-total-2010']) ** (1/12) - 1) * 100
total_population_of_macro_regions_2010_and_2022['taxa-variacao-ano'] = total_population_of_macro_regions_2010_and_2022['taxa-variacao-ano'].round(2)
total_population_of_macro_regions_2010_and_2022
```

```python
total_population_of_macro_regions_projection = total_population_of_macro_regions_2010_and_2022.copy()
```

```python
total_population_of_macro_regions_projection
```

# Creating tables between 2010 and 2022 with the growth rate

```python
total_population_of_macro_regions_projection['taxa-variacao-ano'] = total_population_of_macro_regions_projection['taxa-variacao-ano'] * 0.01


for year in range(2011, 2023):
    column_name = f'populacao-total-{year}'
    previous_year = year - 1
    
    if column_name not in total_population_of_macro_regions_projection.columns:
        total_population_of_macro_regions_projection[column_name] = total_population_of_macro_regions_projection[f'populacao-total-{previous_year}'] * (1 + total_population_of_macro_regions_projection['taxa-variacao-ano'])
total_population_of_macro_regions_projection = total_population_of_macro_regions_projection.applymap(lambda x: str(x)[:8])
pop_2022 = total_population_of_macro_regions_projection.pop('populacao-total-2022')
total_population_of_macro_regions_projection.insert(len(total_population_of_macro_regions_projection.columns), 'populacao-total-2022', pop_2022)
total_population_of_macro_regions_projection = total_population_of_macro_regions_projection.drop('taxa-variacao-ano', axis=1)

total_population_of_macro_regions_projection
```

# Creating csv file


```python
total_population_of_macro_regions_projection.to_csv("total_population_of_macro_regions_projection.csv", index=False)
```
