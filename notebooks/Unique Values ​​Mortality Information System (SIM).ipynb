```python
import pandas as pd
```

```python
df = pd.read_csv("suicidios_2010_a_2019_cleaned.csv")
```

```python
df.head()
```

# Unique Table Values

## State

```python
unique_data_state = df["estado"].unique()
print(unique_data_state)
```

## Race

```python
unique_data_race = df["RACACOR"].unique()
print(unique_data_race)
```

## Year

```python
unique_data_year = df["ano"].unique()
print(unique_data_year)
```

## Month

```python
unique_data_month = df["mes"].unique()
print(unique_data_month)
```

## Gender

```python
unique_data_gender = df["SEXO"].unique()
print(unique_data_gender)
```

## Causes Of Death

```python
unique_data_causes_of_death = df["CAUSABAS"].unique()
print(unique_data_causes_of_death)
```

## Place of occurrence

```python
unique_data_occurrence = df["LOCOCOR"].unique()
print(unique_data_occurrence)
```

## Age

```python
unique_data_age = df["age"].unique()
unique_data_age_sorted = sorted(unique_data_age)
print(unique_data_age_sorted)
```

```python
ages_between_0_and_3_years = df.query('age >= 0 and age <= 3')
ages_between_0_and_3_years
```

```python
ages_between_100_and_113_years = df.query('age >= 100 and age <= 113')
ages_between_100_and_113_years
```
